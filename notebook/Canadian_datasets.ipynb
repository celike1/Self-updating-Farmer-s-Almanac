```python
# custom rounding function to fix adjust lat/lon values to match with CDS agroclimate datasets
def custom_round(num):
    normal_rounded_num = round(num)
    if normal_rounded_num == math.ceil(num):
        return normal_rounded_num - .25
    else:
        return normal_rounded_num + .25
```

```python
print(custom_round(-72.900866))
print(custom_round(48.686761))
print(custom_round(47.347472))
print(custom_round(-47.347472))
```

```python
yield_data['LATITUDE'] = yield_data['LATITUDE'].apply(lambda x: custom_round(x))
yield_data['LONGITUDE'] = yield_data['LONGITUDE'].apply(lambda x: custom_round(x))
yield_data.drop(['CARUID', 'Y2011', 'Y2012', 'Y2013', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018', 'Y2019', 'Y2020', 'Y2021', 'Y2022'], axis=1, inplace=True)
yield_data
```

### Consecutive Wet Days (by season)
Longest period of consecutive days when RR > 1mm, where RR is the daily precipitation sum. This indicator provides information on drought, oxygen stress and crop growth (i.e. less radiation interception during rainy days).

```python
cwd_data = xr.open_dataset(r'C:\Users\kat\Documents\School\2024-WinterB\LSI\CDS_agroclimate\CWD_C3S-glob-agric_gfdl-esm2m_hist_season_19810101-20101231_v1.1.nc')
cwd_data
```

```python
cwd_canada = cwd_data.sel(lat=slice(canada_lat_min, canada_lat_max), lon=slice(canada_lon_min, canada_lon_max))
cwd_canada
```

```python
# map month to seasons
def get_season(month):
    match month:
        case 1:
            return 'winter'
        case 4:
            return 'spring'
        case 7:
            return 'summer'
        case 10:
            return 'fall'
        case _:
            raise Exception('Unintended month input')
```

```python
cwd_df = cwd_canada['CWD'].to_dataframe()
cwd_df.dropna(inplace=True)
cwd_df.reset_index(inplace=True)
cwd_df['YEAR'] = cwd_df['time'].apply(lambda x: x.year)
cwd_df['season'] = cwd_df['time'].apply(lambda x: get_season(x.month))
cwd_df.drop(['time'], axis=1, inplace=True)
print(cwd_df)
cwd_df = pd.pivot_table(cwd_df, index=['YEAR','lat','lon'], columns='season', values='CWD').reset_index()
cwd_df.rename(columns={'spring':'CWD_spring', 'summer':'CWD_summer', 'fall':'CWD_fall', 'winter':'CWD_winter', 'lat':'LATITUDE', 'lon':'LONGITUDE'}, inplace=True)
# winter has a few missing values, also probably not super relevant to crop yields as seeds not sown?
cwd_df.drop(['CWD_winter'], axis=1, inplace=True)
cwd_df
```

```python
cleaned_data = cleaned_data.merge(cwd_df, on=['YEAR', 'LATITUDE', 'LONGITUDE'])
cleaned_data
```

### Warm and Wet Days (by season)
Number of days per 10 days when TG > TG75th and RR > RR75th; where TG is the daily mean temperature, TG75th is the calendar day 75th percentile, RR is the daily precipitation sum and RR75th is the 75th percentile of precipitation on wet days. To calculate TG75th and RR75 for each day, a 5-day window centered on each equivalent day in the 1981-2010 period is used to improve the statistical robustness from which the 75th percentile of the data is selected (e.g. for all the 3rd of July days between 1981-2010, the 1st, 2nd, 3rd, 4th, 5th July are selected in the 5-day window). This indicator provides an indication of occurrence of various pests and the crop development, especially leaf formation.
